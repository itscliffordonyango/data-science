```python
import pandas as pd 
dataset = pd.DataFrame({
                        "Weight":[65,68,52,79,55,60,71,55,85,100],
                        "Height":[180,176,195,210,155,168,170,188,150,145]
                        })
print(dataset)
```

Models to implement regression , scikit-learn

y = B0 + B1x

```python
import pandas as Pd
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# dataset = pd.DataFrame({
#                         "Weight":[65,68,52,79,55,60,71,55,85,100],
#                         "Height":[180,176,195,210,155,168,170,188,150,145]
#                         })


Weight=Pd.DataFrame(dataset["Weight"])
Height =Pd.DataFrame(dataset["Height"])

model.fit(Weight,Height)
```

```python
pred_values = model.predict(Weight)
print(pred_values)
```

```python
model.coef_  # B1
```

```python
model.intercept_  # B0
```

```python
y = model.intercept_ + model.coef_*Weight
print(y)
```

```python
data = pd.read_csv("BostonHousing.csv")
data.head()
```

```python
data.shape
```

Laso regression 

```python
from sklearn import linear_model
from sklearn.linear_model import Lasso

lasso_reg = Lasso()

from sklearn.model_selection import train_test_split

depVarriable = "MEDV"
indepVarriables = data.columns.tolist()
indepVarriables.remove(depVarriable)
indepVarriables
```

```python
depVarriable
```

```python
from sklearn import datasets
# datasets.load_boston()


x = data[indepVarriables].values
y = data[depVarriable].values

print(x)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
lasso_reg.fit(X_train, y_train)
```

```python
predictions = lasso_reg.predict(X_test)
print(predictions)
```

```python
coeffs = lasso_reg.coef_
print(coeffs)
```

```python
lasso_reg.intercept_
```

```python
from sklearn.metrics import mean_squared_error, r2_score
r2_score(y_test, predictions)
```

ridge regression

```python
from sklearn import linear_model
from sklearn.linear_model import Ridge

ridge_reg = Ridge()
ridge_reg.fit(X_train, y_train)

```

```python
predictions = ridge_reg.predict(X_test)
print(predictions)
```

```python
intercept = ridge_reg.intercept_
print(intercept)
```

```python
r2_score(y_test, predictions)
```
