```python
import pandas as pd

data = "train.csv"

dataset = pd.read_csv(data)
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
# check for missing values
dataset.isnull().sum()
# fill rows with missing values
dataset.Age.fillna(method='ffill')
dataset.isnull().sum()
dataset.head()
```

```python
dataset.Embarked = dataset.Embarked.fillna(method='ffill')
```

```python
dataset.isnull().sum()
```

```python
# check for duplicate rows  
dataset.duplicated().sum()

```

```python
# drop the cabin 

dataset.drop("Cabin", axis=1, inplace=True)
dataset.drop("Ticket", axis=1, inplace=True)
dataset.drop("Name", axis=1, inplace=True)
```

```python
dataset
```

```python
dataset.value_counts(dataset.Embarked)
```

```python
def tran_mathScore(x):
    if x == "s" :
        return 0
    if x == "C":
        return 1
    if x == "Q":
        return 2
    
```

```python
dataset.Embarked = dataset.Embarked.apply(tran_mathScore)
dataset.Sex = dataset.Sex.apply(lambda x: 0 if x == "male" else 1)  
```

```python
# dataset.drop("Cabin", axis=1, inplace=True)
dataset
```

```python
from sklearn.model_selection import train_test_split

# Dependent (target) variable
dependent_variable = "Survived"

# Independent variables
independent_variable = dataset.columns.to_list()
independent_variable.remove(dependent_variable)

# Separate X (features) and y (target)
x = dataset[independent_variable].values
y = dataset[dependent_variable].values

# Check shapes
print("Shape of X:", x.shape)
print("Shape of y:", y.shape)

# Split into train/test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("\nTraining set shape:", x_train.shape)
print("Test set shape:", x_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
predict = dtree.predict(x_test)
predict
```
