```python
data  = {"waist":[71,89,64,74,87,93,79,81,75,72,7066,81,78,110],"weight":[51,66,45,57,64,75,61,62,55,5348,46,63,60,100]}
import matplotlib.pyplot as plt
plt.scatter(data["waist"], data["weight"])
plt.xlabel("Waist (cm)")
plt.ylabel("Weight (kg)")
plt.title("Waist vs Weight")
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
X = np.array(data["waist"]).reshape(-1, 1)
y = np.array(data["weight"])
model = DecisionTreeRegressor()
model.fit(X, y)
predictions = model.predict(X)
plt.scatter(data["waist"], data["weight"])
plt.plot(data["waist"], predictions, color='red')
plt.xlabel("Waist (cm)")
plt.ylabel("Weight (kg)")
plt.title("Decision Tree Regression: Waist vs Weight")
```

```python
import scipy.stats as stats
# lets determine the regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(data["waist"], data["weight"])
line = [slope * xi + intercept for xi in data["waist"]]
plt.scatter(data["waist"], data["weight"])
plt.plot(data["waist"], line, color='green')
plt.xlabel("Waist (cm)")
plt.ylabel("Weight (kg)")
plt.title("Linear Regression: Waist vs Weight")
```

```python
# correlation coefficient
correlation_matrix = np.corrcoef(data["waist"], data["weight"])
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2
print(f"R-squared: {r_squared}")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Your dataset
data = {
    "waist": [71, 89, 64, 74, 87, 93, 79, 81, 75, 72, 70, 66, 81, 78, 110],
    "weight": [51, 66, 45, 57, 64, 75, 61, 62, 55, 53, 48, 46, 63, 60, 100]
}

# Create DataFrame
df = pd.DataFrame(data)

# Train a simple linear regression model
X = df["waist"].values.reshape(-1, 1)
y = df["weight"].values

model = LinearRegression()
model.fit(X, y)

# Make predictions
pred = model.predict(X)

# Combine actual vs predicted
weight = pd.DataFrame(y, columns=["Actual Weight"])
pred = pd.DataFrame(pred, columns=["Predicted Weight"])

output = pd.concat([weight, pred], axis=1)
print(output)

# Optional: summary statistics


```

```python
weight = pd.DataFrame(data["weight"], columns=["weight"])
weight.describe()
pred = pd.DataFrame(pred)
output = pd.concat([weight, pred], axis=1)
output.columns = ["Actual Weight", "Predicted Weight"]
output
```

```python
# plot the regression line 
import matplotlib.pyplot as plt
plt.scatter(data["waist"], data["weight"])
waist = np.array(data["waist"]).reshape(-1, 1)
pred = model.predict(waist)
plt.plot(waist, pred, color='red')
plt.xlabel("Waist (cm)")
plt.ylabel("Weight (kg)")
plt.title("Linear Regression: Waist vs Weight")
```

```python
from sklearn.metrics import mean_squared_error, r2_score
r2 = r2_score(y, pred)
print(f"R-squared: {r2}")
```
